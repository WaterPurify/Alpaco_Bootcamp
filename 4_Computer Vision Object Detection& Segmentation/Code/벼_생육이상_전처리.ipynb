# 파일 이동

```python
def move_files(src_dir, dest_dir):

    # Get list of files in the source directory
    file_list = [f for f in os.listdir(src_dir) if os.path.isfile(os.path.join(src_dir, f))]

    for file in tqdm.tqdm(file_list):
        src_path = os.path.join(src_dir, file)
        dest_path = os.path.join(dest_dir, file)
        shutil.move(src_path, dest_path)
```

```python
def img_compose(
    data_path: str, channels=["E", "N", "G"], mode="chw_minmax", replace=False) -> str:
    """Compose png images from channel-wise data (mostly in .tif format)

    Args:
        data_path (str): the path of the 'data' directory
        channels (list, optional): Among ["R", "G", "B", "N", "E], select three (repetition allowed). Defaults to ["E", "N", "G"].
        mode (str, optional): float32 to int8 convert and normalize method. Defaults to 'chw_minmax'.
        replace (bool, optional): replace existing results or not. Defaults to False.

    Raises:
        NotImplementedError: when the mode name is not supported

    Returns:
        str: img_conf_name
    """
    print("Composing images...")
    assert len(channels) == 3
    img_dir = os.path.join(data_path, "img")         # C:\Users\campus1N029\Desktop\data\img
    img_conf_name = "".join(channels) + "_" + mode   # ENG_chw_minmax
    output_dir = os.path.join(data_path, img_conf_name)
    if not replace:
        if os.path.exists(output_dir):
            print(
                f"For image composition, using existing img_dir: {os.path.abspath(output_dir)}"
            )
            return img_conf_name
    os.makedirs(output_dir, exist_ok=True)

    if mode == "chw_minmax":
        # Channel-wise minmax normalization
        for channel in channels:
            for file in tqdm.tqdm(sorted(os.listdir(os.path.join(img_dir, channel)))):   # C:\Users\campus1N029\Desktop\data\img\E
                img = []
                #try:
                    # for ch_id, channel in enumerate(channels):
                cur_chn = np.array(Image.open(os.path.join(img_dir, channel, file)))  
                cur_chn[cur_chn < 0] = cur_chn[cur_chn != -10000].min()
                cur_chn = (cur_chn - cur_chn.min()) / cur_chn.ptp()
                img.append(cur_chn)
                img = np.stack(img, axis=2)
                cv2.imwrite(os.path.join(output_dir, file[:-3] + "png"), img * 255)
                #except:
                    #os.remove(os.path.join(data_path, "cls_ann", file[:-3] + "png"))
    elif mode == "minmax":
        # minmax normalization
        for file in tqdm.tqdm(sorted(os.listdir(os.path.join(img_dir, channels[0])))):
            img = []
            for ch_id, channel in enumerate(channels):
                cur_chn = np.array(Image.open(os.path.join(img_dir, channel, file)))
                img.append(cur_chn)
            img = np.stack(img, axis=2)
            img[img == -10000] = np.min(img[img != -10000])
            img = (img - img.min()) / img.ptp()
            cv2.imwrite(os.path.join(output_dir, file[:-3] + "png"), img * 255)
            
    elif mode == "cv2_merge":
        # cv2.merge automatically normalize

        # G_list = []
        # N_list = []
        # E_list = []
    
        # for channel in channels:
        #     for file in tqdm.tqdm(sorted(os.listdir(os.path.join(img_dir, channel)))):

        #         if file[-9] == 'G':
        #             org_cur_chn = cv2.imread(
        #                 os.path.join(img_dir, channel, file), cv2.IMREAD_UNCHANGED
        #             )
        #             G_list.append(org_cur_chn)
        #             print(len(G_list))
        #         elif file[-9] == 'N':
        #             org_cur_chn = cv2.imread(
        #             os.path.join(img_dir, channel, file), cv2.IMREAD_UNCHANGED
        #             )
        #             N_list.append(org_cur_chn)
        #             print(len(N_list))
        #         else:
        #             org_cur_chn = cv2.imread(
        #             os.path.join(img_dir, channel, file), cv2.IMREAD_UNCHANGED
        #             )
        #             E_list.append(org_cur_chn)
        #             print(len(E_list))

        # for i in range(len(G_list)):               
        #     img = cv2.merge([G_list[i],N_list[i],E_list[i]])
        #     cv2.imwrite(os.path.join(output_dir, file[:-3] + "png"), img * 255)
        # else:
        #     raise NotImplementedError
        E_file = 'C:/Users/campus1N029/Desktop/data/img/E'
        G_file = 'C:/Users/campus1N029/Desktop/data/img/G'
        N_file = 'C:/Users/campus1N029/Desktop/data/img/N'

        G_files = sorted(os.listdir(G_file))
        N_files = sorted(os.listdir(N_file))
        E_files = sorted(os.listdir(E_file))

        for i in tqdm.tqdm(range(len(os.listdir(G_file)))):
            # Get the file names
            G_filename = G_files[i]
            N_filename = N_files[i]
            E_filename = E_files[i]
            
            # Construct the full file paths
            G_path = os.path.join(G_file, G_filename)
            N_path = os.path.join(N_file, N_filename)
            E_path = os.path.join(E_file, E_filename)

            # Read the images
            G_img = cv2.imread(G_path, cv2.IMREAD_UNCHANGED)
            N_img = cv2.imread(N_path, cv2.IMREAD_UNCHANGED)
            E_img = cv2.imread(E_path, cv2.IMREAD_UNCHANGED)

            # Merge the images
            img = cv2.merge([G_img, N_img, E_img])

            # Write the image to the output directory
            output_filename = os.path.join(output_dir, G_filename[:-3] + "png")
            cv2.imwrite(output_filename, img* 255)

    print("Image composition is done!")
    return img_conf_name
```

```python
def move_files(src_dir, dest_dir):

    dic = src_dir
    file_list = os.listdir(dic)

    for file in tqdm.tqdm(file_list):
        if file[-10] == 'G':
            src_path = os.path.join(src_dir, file)
            dest_path = os.path.join(dest_dir, file)
            shutil.move(src_path, dest_path)
```

```python
def code2cls(obj_cls_code):
    if obj_cls_code == "01":
        cls = 1
    elif obj_cls_code == "02":
        cls = 2
    elif obj_cls_code == "03":
        cls = 3
    elif obj_cls_code == "04":
        cls = 4
    elif obj_cls_code == "05":
        cls = 5
    else:
        cls = 0
    return cls
```
